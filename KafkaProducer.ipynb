```python
import pandas as pd
from kafka import KafkaProducer
from time import sleep
from json import dumps
import json
```

```python
producer = KafkaProducer(bootstrap_servers=[''], #change ip here
                         value_serializer=lambda x:
                         dumps(x).encode('utf-8'))
```

```python
producer.send('KafkaStockMarket', value={'Full Name':'Muhammad Shaheer Jamal Chishti'})
```

```python
df = pd.read_csv("./indexProcessed.csv")
```

```python
# df.head()
# df.sample()
df.sample().to_dict(orient="records")[0]
```

```python
while True:
    dict_stock = df.sample(1).to_dict(orient="records")[0]
    producer.send('KafkaStockMarket', value=dict_stock)
    sleep(0)
```

```python
producer.flush() #clear data from kafka server
```
